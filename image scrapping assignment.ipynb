```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException  
from selenium.common.exceptions import SessionNotCreatedException
import time
import selenium
from PIL import Image
import pandas as pd
import urllib
from selenium import webdriver
```

```python
driver=webdriver.Chrome('chromedriver.exe')
```

```python
urls = []

base_url = "https://www.amazon.in/s?k=Sarees&page={}&qid=1632112444&ref=sr_pg_{}"
driver.get(base_url)
```

```python
for i in range(2,10):
    
    driver.get(base_url.format(i,i))
    
    for url in driver.find_elements_by_xpath("//a[@class='a-link-normal s-no-outline']"):
        
        urls.append(url.get_attribute("href"))
```

```python
urls = urls[:300]
len(urls)
```

```python
# Grabbing image source

images_sarees = []

for link in urls:
    
    driver.get(link)
    time.sleep(1)
    
    try:
        
        img = driver.find_element_by_css_selector("#landingImage")
    
        source= img.get_attribute('src')
    
        images_sarees.append(source)
        
    except NoSuchElementException:
        
        images_sarees.append("Image Not Found!") 
```

```python
import requests


for i in range(len(images_sarees)):
    
    print("Downloading....")
    
    response= requests.get(images_sarees[i])
    
    file = open(r"D:\30-Nov-2021\Sarees"+str(i)+".jpg", "wb")
    
    file.write(response.content)
```
